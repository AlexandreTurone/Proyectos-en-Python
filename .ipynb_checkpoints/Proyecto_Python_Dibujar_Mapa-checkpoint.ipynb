```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from netCDF4 import Dataset as nc
import matplotlib.dates as mdates
import datetime 


url = '/Users/alexandremartinez/Downloads/train.nc'

def data_r(url):
    file = nc(url, 'r') #Leemos el fichero NC
    print('¿Desea ver el archivo?')
    x = input('¿si o no?')
    if x == 'si':
        print(file)
     
    
def ciudad(lat_real, lon_real):
    
    lat = input('Ponga el nombre de la variable latitude:')
    lon = input('Ponga el nombre de la variable longitude:')
    
    latitude = file.variables[str(lat)][:]
    longitude = file.variables[str(lon)][:]

   
    
    for i in range(len(latitude)):
        if latitude[i]<=(lat_real + 0.5) and latitude[i]>=(lat_real - 0.5):
            print('Indice:',i,'Latitudes:',latitude[i])
        
        
        
    for i in range(len(longitude)):
        if longitude[i]<=(lon_real + 0.5) and longitude[i]>=(lon_real - 0.5):
            print('indice:', i, 'Longitudes:', longitude[i])
    
    print('Elija índice asociado a la mejor latitude y longitude:')
    ind_lat = input('Indice Latitude:')
    ind_lon = input('Indice Longitude:')
    

    
    
    return ind_lat, ind_lon







```

```python
url = '/Users/alexandremartinez/Downloads/train.nc'
file = nc(url, 'r') #Leemos el fichero NC

print(file.variables)
```

```python
lat = file.variables['latitude'][:]
lon = file.variables['longitude'][:]
rocio = file.variables['d2m'][:]
temp = file.variables['t2m'][:]
press = file.variables['sp'][:]
precip = file.variables['tp'][:]
tiempo = file.variables['time'][:]
cloud = file.variables['cbh'][:]
ssr = file.variables['ssr'][:]
```

```python
ciudad( 37.98, -1.13 ) #Murcia
```

```python
indice_lat = 24
indice_lon = 36

precip_ciudad = precip[:,0,indice_lat,indice_lon]*1000
temp_ciudad = temp[:,0,indice_lat,indice_lon] - 273.15
press_ciudad = press[:,0,indice_lat,indice_lon]*0.01
d2m_ciudad = rocio[:,0,indice_lat,indice_lon]-273.15
cloud_ciudad = cloud[:,0,indice_lat,indice_lon]
ssr_ciudad = ssr[:,0,indice_lat,indice_lon]

# Definimos las fechas del dataset
start = datetime.datetime(2022,1,1,0,0)
periods = len(precip)
daterange = []
for day in range(periods):
    date = (start + datetime.timedelta(hours = day)).isoformat()
    daterange.append(date)

        
dataset = pd.DataFrame()
dataset['Precip']= precip_ciudad
dataset['Temp']= temp_ciudad
dataset['Press']= press_ciudad
dataset['Tempe.Rocio']= d2m_ciudad
dataset['cloud'] = cloud_ciudad
dataset['rad'] = ssr_ciudad
dataset.index = daterange
dataset.head()

print('Longitud del dataset antes:', len(dataset))

dataset = dataset.dropna()

print('Longitud del dataset despues:', len(dataset))


dataset.head()

```
